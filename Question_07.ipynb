# __Question 7__

Create a contour plot for the function 
**Z = X ** 2 + Y ** 2** where X,Y 
are created using meshgrid(x,y) for the following points:  
```bash 
x= np.linspace(-5,5,100) 
y= np.linspace(-5,5,100)
```

````python
# Importing neccessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Generate linearly spaced values for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a mesh grid
X, Y = np.meshgrid(x, y)

# Define the function Z = np.sin(X) + np.cos(Y)
Z = np.sin(X) + np.cos(Y)

# Create a contour plot
contour_plot = plt.contour(X, Y, Z)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Contour Plot")

# Add a color bar
plt.colorbar(contour_plot, label="Z")

# Display the plot
plt.show()
````
