````python
# Importing neccessary libraries.
import matplotlib.pyplot as plt 
import numpy as np

# Create a 3D axis
ax = plt.axes(projection="3d")

# Generate data
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

# Plot the 3D line
ax.plot3D(x, y, z, "hotpink")

# Add a title
ax.set_title("3D - Line Plotting")

# Display the plot
plt.show()
````
